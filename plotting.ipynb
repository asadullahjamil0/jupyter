```python
import numpy as np
X = np.linspace(0 ,10, 40)
print (X)
```

```python
C = np.cos(X)
print(C)
```

```python
import matplotlib.pyplot as plt
```

%matplotlib inline

```python
plt.plot(X, C)
```

```python
S = np.sin(X)
print(S)
```

```python
plt.figure()
plt.plot(X, C)
plt.plot(X, S)
```

```python
plt.figure()
plt.plot(X, C, label="cos")
plt.plot(X, S, label="sin")
plt.xlabel('x')
plt.ylabel('sin(x)/cos(y)')
plt.legend()
```
